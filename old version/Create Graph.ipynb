```python
import networkx as nx
import json
import matplotlib.pyplot as plt
```

```python
d = {}
with open("Similar Words.json","r+") as f:
    d = json.load(f)
```

```python
languages = []
with open("All_Languages.txt","r+") as f:
    languages = f.read().split("\n")
```

```python
graph = nx.Graph()
```

```python
for i in languages:
    graph.add_node(i)
```

```python
x = []
for key,value in d.items():
    for i in value:
        graph.add_edge(key.lower(),i.lower())
    for i in value:
        for j in value:
            if i!=j:
                graph.add_edge(i.lower(),j.lower())
        if i.lower() not in languages:
            x.append(i.lower())
```

```python
plt.figure(figsize=(20,15))
nx.draw(graph,with_labels=True)
```

```python
for i in graph.neighbors("react"):
    print(i+"-->")
    for j in graph.neighbors(i):
        print(j,end=" ")
    print()
```
